```python
import pandas as pd

# Load the dataset
df = pd.read_csv('running_with_binary_fatigue.csv')

# View the first few rows
df.head()
```

```python
# Check basic info
df.info()

# See how many entries are fatigued vs not
df['fatigue'].value_counts()
```

```python
# Check for missing values
print(df.isnull().sum())

# Fill missing values (e.g., with mean for numerical columns)
df.fillna(df.mean(), inplace=True)

# Alternatively, drop rows with missing values if they are few
# df.dropna(inplace=True)
```

```python
# Check data types of each column
print(df.dtypes)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the Label Encoder
label_encoder = LabelEncoder()

# Apply Label Encoding to categorical columns
categorical_columns = ['gender', 'activity_type', 'intensity']
for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])

# Check the updated dataset
print(df.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Separate features (X) and target (y)
X = df.drop('fatigue', axis=1)  # Features (all columns except 'fatigue')
y = df['fatigue']              # Target ('fatigue')

# Scale the features using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Check the scaled features
print("First 5 rows of scaled features:")
print(X_scaled[:5])
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Check the sizes of the training and testing sets
print("Training set size (X_train):", X_train.shape)
print("Testing set size (X_test):", X_test.shape)
print("Training labels size (y_train):", y_train.shape)
print("Testing labels size (y_test):", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

# Initialize the Logistic Regression model
log_reg = LogisticRegression(max_iter=1000, random_state=42)

# Train the model on the training data
log_reg.fit(X_train, y_train)

# Predict on the test set
y_pred_log_reg = log_reg.predict(X_test)

# Evaluate the model's accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred_log_reg)
print("Logistic Regression Accuracy:", accuracy)
```

```python
# Check the distribution of the 'fatigue' column
print(df['fatigue'].value_counts())
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Generate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_log_reg)
print("Confusion Matrix:")
print(conf_matrix)

# Generate the classification report
class_report = classification_report(y_test, y_pred_log_reg)
print("\nClassification Report:")
print(class_report)
```
